#### Números

```python
print(2) # Enteros, sin parte fraccional.
print(3.4) # Números en coma flotante, con parte fraccional.
print(2+4j) # Números complejos.
print(1/2) # Numeros racionales.
```

#### Strings
- Cadenas de caracteres.
- Son secuencias: la posición de los caracteres es importante.
- Son immutables: las operaciones sobre strings no cambian el string original.

```python
s = 'John "ee" Doe'
print(s[0]) # Primer carácter del string.
print(s[-1]) # Último carácter del string.
print(s[1:8:2]) # Substring desde el segundo carácter (inclusive) hasta el octavo (exclusive). Esta técnica se la conoce como 'slicing'.
print(s[:]) # Todo el string.
print(s + "e") # Concatenación.
```

### 2.4 Conversión entre tipos
- A veces se quiere que un objeto sea de un tipo específico.
- Podemos obtener objetos de un tipo a partir de objetos de un tipo diferente (casting)

```python
a = int(2.8) # a será 2
b = int("3") # b será 3
c = float(1) # c será 1.0
d = float("3") # d será 3.0
e = str(2) # e será '2'
f = str(3.0) # f será '3.0'
g = bool("a") # g será True
h = bool("") # h será False
i = bool(3) # i será True
j = bool(0) # j será False
k = bool(None)
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
print(e)
print(type(e))
print(f)
print(type(f))
print(g)
print(type(g))
print(h)
print(type(h))
print(i)
print(type(i))
print(j)
print(type(j))
print(k)
```

```python
print(7/4) # División convencional. Resultado de tipo 'float'
print(7//4) # División entera. Resultado de tipo 'int'
print(int(7/4)) # Divisón convencional. Conversión del resultado de 'float' a'int'
```

### 2.5 Operadores
- Combinación de valores, variables y operadores
- Operadores y operandos
#### Operatores aritméticos
----------------
Operador| Desc
<br>

<br>

<br>

```python
x = 3
y = 2
print('x + y = ', x + y)
print('x - y = ', x - y)
print('x * y = ', x * y)
print('x / y = ', x / y)
print('x // y = ', x // y)
print('x % y = ', x % y)
print('x ** y = ', x ** y)
```

#### Operadores Lógicos

----------------
Operador| Desc
<br>
a and b  | True, si ambos son True
<br>
a or b |   True, si alguno de los dos son True
<br>
a ^ b |   XOR- True, si solo uno de los dos es True
<br>
not a|  Negación


```python
x = True
y = False
print('x and y es :', x and y)
print('x or y es :', x or y)
print('x xor y es :', x ^ y)
print('not x es :', not x)
```

#### Operadores Binarios
----------------

<br>

<br>
a % b |  Modulo / Resto
<br>

```python
x = 0b01100110
y = 0b00110011
print("Not x = " + bin(~x))
print("x and y = " + bin(x & y))
print("x or y = " + bin(x | y))
print("x xor y = " + bin(x ^ y))
print("x << 2 = " + bin(x << 2))
print("x >> 2 = " + bin(x >> 2))
```

### 3.1 Listas
- Una colección de objetos.
- Mutables.
- Tipos arbitrarios heterogeneos.
- Puede contener duplicados.
- No tienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.
- Los elementos van ordenados por posición.
- Se acceden usando la sintaxis.
- Los índices van de 0 a n-1, donde n es el número de elementos de la lista.
- Son un tipo de Secuencia, al igual que los strings; por lo tanto, el orden (es decir, la posición de los objetos de la lista) es importante.
- Soportan anidamiento.
- Son una implementación del tipo abstracto de datos: Array Dinámico.
### Operaciones con listas

```python
letras = ['a', 'b', 'c', 'd']
palabras = 'Hola mundo como estas'.split()
numeros = list(range(7))
print(letras)
print(palabras)
print(numeros)
print(type(numeros))
```

```python
# # Pueden contener duplicados
lista_con_duplicados = [1, 2, 3, 3, 3, 4]
print(lista_con_duplicados)
```

- Alias/Referencias en las listas
- Son mutables y las asignaciones a un objeto alteran el primero
- Pasar listas a funciones puede suponer un riesgo
- Clonar o copiar listas

```python
lista = [2, 4, 16, 32]
ref = lista
print(id(lista))
print(id(ref))
ref[2] = 64
print(ref)
print(lista)
```

```python
lista = [2000, 4, 16, 32]
copia = lista[:]
copia = lista.copy()
copia[2] = 64
print(lista)
print(copia)
print(id(lista))
print(id(copia))
print(id(lista[2]))
print(id(copia[2]))
```

```python
lista = [0, 1, [10, 20]]
print(id(lista))
lista2 = [10, 20]
lista = [0, 1, lista2]
copia = lista[:] #.copy()
print(copia)
print(id(copia))
print(id(lista[2]))
print(id(copia[2]))
copia[2][0] = 40
print(copia)
print(lista)
```

```python
# evitar que listas anidadas se copien por referencia
import copy
lista = [2, 4, 16, 32, [34, 10, [5,5]]]
copia = copy.deepcopy(lista)
copia[0] = 454
copia[4][2][0] = 64
print(lista)
print(copia)
print(f"{id(lista)} - {id(copia)}")
print(f"{id(lista[4])} - {id(copia[4])}")
```

### 3.2 Diccionarios
- Mapeo clave→valor, mutable, ordenado (Py ≥3.7).
- Claves: inmutables y únicas.
- Valores: cualquier objeto.

#### Operaciones de diccionarios
- Creacion de diccionarios

```python
# Creación simple, usando una expresión literal.
persona = {'DNI' : '11111111D', 'Nombre' : 'Juan', 'Edad' : 26}
print(persona)
```

```python
# Creación uniendo dos colecciones.
nombres = ['Roberto', 'Jorge', 'Andrés', 'Carlos']
edades = [52, 14, 65]
datos = dict(zip(nombres,edades))
print(datos)
```

```python
# Creación pasando claves y valores a la función 'dict'
persona2 = dict(nombre='Ana', apellido='Lopez')
print(persona2)
```

```python
# Creación usando una lista de tuplas de dos elementos.
persona2 = dict([('nombre', 'Ana'), ('apellido','Lopez')])
print(persona2)
```

```python
## Creación incremental por medio de asignación (como las claves no existen, se crean nuevos items)
persona = {}
# print(persona['C.I'])
persona['C.I'] = '11111111D'
persona['Nombre'] = 'Juan'
persona['Edad'] = 26
persona['C.I'] = '222222222'
print(persona)
```

- Acceso a un valor  con la clave

```python
persona = {'C:I' : '11111111D', 'Nombre' : 'Juan', 'Edad' : 26}
print(persona['Nombre'])
```

```python
# Acceso a claves inexistentes o por índice produce error
# persona[1]
print(persona['Nombre'])
if 'Trabajo' in persona:
print(persona['Trabajo'])
else:
print("En el paro")
persona.get('Trabajo', "En el paro")
```

- Modificación a un valor usando la  clave

```python
persona = {'DNI' : '11111111D', 'Nombre' : 'Juan', 'Edad' : 26}
persona['Nombre'] = 'Alejandro'
persona['Edad'] += 1
print(persona)
```

- Agregar un valos usando la clave

```python
persona = {'C.I' : '11111111D', 'Nombre' : 'Juan', 'Edad' : 26}
persona['Ciudad'] = 'Madrid'
print(persona)
```

- Eliminar un valor con la clave

```python
del persona['Ciudad']
value = persona.pop('Edad')
print(persona)
print(value)
```

- Verificar existencia de clave

```python
print('Nombre' in persona)
print('Apellido' in persona)
```

- Recuperar un valor de un clave, si no existe indica valor por defecto

```python
persona = {'Nombre' : 'Jose'}
value = persona.get('Nombre')
print(value)
# persona['Estatura']
value = persona.get('Estatura', 180)
print(value)
```

- Anidamiento

- Métodos  (Items, claves, valores)

```python
persona = {'DNI' : '11111111D', 'Nombre' : 'Juan', 'Edad' : 26}
print(list(persona.items()))
print(list(persona.keys()))
print(list(persona.values()))
```

```python
dict_simple = {'ID' : 'XCSDe1194', 'Nombre' : 'Juan', 'Edad' : 26}
# iterar sobre las claves
print('Claves')
for key in dict_simple.keys(): # o dict_simple
    print(key)
print('\nValores')
# iterar sobre los valores
for value in dict_simple.values():
    print(value)
```

```python
dict_simple = {'ID' : 'XCSDe1194', 'Nombre' : 'Juan', 'Edad' : 26}
# iterar sobre ambos, directamente con items().
for key, value in dict_simple.items():
    print('Clave: ' + str(key) + ', valor: ' + str(value))
for item in dict_simple.items(): # devuelve tuplas
    print('Item: ' + str(item))
# usar zip para iterar sobre Clave-Valor
for key, value in zip(dict_simple.keys(),dict_simple.values()):
    print('Clave: ' + str(key) + ', Valor: ' + str(value))
```

### 3.3 Conjuntos
- Colección mutable, sin duplicados, sin orden.
- Solo elementos hashables (inmutables).
- Útiles para álgebra de conjuntos.

#### Operaciones con conjuntos
- Creacin de conjuntos

```python
set_1 = {0, 1, 1, 2, 3, 4, 4}
print(set_1)
set_2 = {'usuario1', 12, 2}
print(set_2)
set_3 = set(range(7))
print(set_3)
set_4 = set([0, 1, 2, 3, 4, 0, 1])
print(set_4)
```

```python
#Observa la diferencia entre listas y conjuntos
x = 'aabbc'
print(list(x))
print(set(x))
```

- Intenar acceder por indice puede causar errores

```python
set1 = {0, 1, 2}
print(set1[0])
```

- Unión, intersección y diferencia

```python
set_1 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set_2 = set([0, 1, 2, 3, 4, 42])
# union
print(set_1 | set_2)
# intersección
print(set_1 & set_2)
# diferencia
print(set_1 - set_2)
print(set_2 - set_1)
```

```python
x = [1,2]
y = [2,3]
print(set(x) & set(y))
```

```python
#Además de los operadores, que operan únicamente con Sets, también se pueden usar métodos que pueden operar sobre cualquier objeto iterable.
conjunto = {0, 4, 8}
lista = [2, 6, 9]
print(conjunto.union(lista))
print(conjunto.intersection(lista))
print(conjunto.difference(lista))
```

- Comparación

```python
set3 = {0, 1, 1, 2, 3, 4, 5, 8, 13, 21}
set4 = set([0, 1, 2, 3, 4])
print(set4.issubset(set3))
print(set3.issuperset(set4))
print(set3.isdisjoint(set4))
```

- Pertenencia

```python
words = {'calm', 'balm'}
print('calm' in words)
```

- Anidamiento

```python
# Los conjuntos no soportan anidamiento, pero como permite elementos inmutables, se pueden "anidar" tuplas.
set_anidado = {1, (1, 1, 1), 2, 3, (1,1,1)}
print(set_anidado)
```

- Modificaciones de conjuntos

```python
# A través de operador de asignación
set6 = {'a', 'b', 'c'}
set7 = {'a', 'd'}
# set6 |= set7 # set6 = set6 | set7
# set6 &= set7
set6 -= set7
print(set6)
```

```python
# A través de método 'update'.
set_1 = {'a', 'b', 'c'}
set_2 = {'a', 'd'}
# set_1.update(set_2)
# set_1.intersection_update(set_2)
set_1.difference_update(set_2)
print(set_1)
```

```python
# A través de métodos 'add' y 'remove'.
set5 = {'a', 'b', 'c'}
set5.add('d')
set5.remove('a')
print(set5)
```

### 3.4 Tuplas
- Secuencia inmutable.
- Pueden usarse como claves de dict.
- Desempaquetado y slicing idénticos a listas.


#### Operaciones con tulpas
- Creación de tuplas.

```python
tuple_1 = ('Foo', 34, 5.0, 34)
print(tuple_1)
tuple_2 = 1, 2, 3
print(tuple_2)
tuple_3 = tuple(range(10))
print(tuple_3)
tuple_4 = tuple([0, 1, 2, 3, 4])
print(tuple_4)
```

```python
singleton = (99,)
print(singleton, type(singleton))
```

```python
print(len(tuple_1))
```

- Acceso por indice

```python
tupla1 = ('Foo', 1, 2, 3)
print(tupla1[0]) # Primer elemento
print(tupla1[len(tupla1)-1]) # Último elemento
print(tupla1[-1]) # Índices negativos comienzan desde el final
print(tupla1[-len(tupla1)]) # primer elemento
```

- Asignarcion a tuplas provoca errores

```python
tupla1[0] ='bar'
```

- Contar numero de repeticiones de un elemento

```python
tuple1 = ('Foo',34, 5.0, 34)
print(tuple1.count(34))
```

- Encontrar indice d eun elemento

```python
tuple1 = ('Foo', 34, 5.0, 34)
indice = tuple1.index(34)
print(indice)
print(tuple1[indice])
```

- Desempaquetar una tupla

```python
tuple1 = (1, 2, 3, 4)
a, b, c, d = tuple1 # a,b,c - a,b,c,d - a,b,_,_ - a, *_
print(a)
print(b)
print(_)
print(d)
```

|Funcion| Uso |

|enumerate|	for i, v in enumerate(lista):|
|zip	|Une elemento a elemento; se detiene al más corto.|


```python
list(range(0,10,2))
```

```python
lista = [10, 20,30]
print(list(enumerate(lista)))
```

```python
# enumerate para iterar una colección (índice y valor)
lista = [10, 20, 30]
for index, value in enumerate(lista):
    print('Index = ' + str(index) + '. Value = ' + str(value))
for index, value in [(0,10), (1,20), (2,30)]:
    print('Index = ' + str(index) + '. Value = ' + str(value))
```

```python
# enumerate para iterar una colección (índice y valor)
for index, value in enumerate(range(0,10,2)): 
    print('Index = ' + str(index) + '. Value = ' + str(value))
for value in enumerate(range(0,10,2)):
    print(value)
```

```python
# zip para unir, elemento a elemento, dos colecciones, retornando lista de tuplas
# útil para iterar dos listas al mismo tiempo
nombres = ['Fernando','Gabriel','Diego']
edades = [21,24,23,22]
for nombre,edad in zip(nombres,edades): 
    print('Nombre: ' + nombre + ', edad: ' + str(edad))
```

```python
nombres = ['Fernando','Gabriel','Diego']
edades = [21,24,23,22]
for i in range(0,len(nombres)):
    print(f"Nombre es {nombres[i]} y edad es {edades[i]}")
```

```python
nombres = ['Fernando','Gabriel','Diego']
edades = [21,24,23,22]
jugadores = zip(nombres,edades) # genera secuencia
print(list(jugadores))
print(type(jugadores))
```

```python
nombres = ['Fernando','Gabriel','Diego']
edades = [21,24,23,22]
jugadores = zip(nombres,edades)
print(list(jugadores))
```

```python
#unzip
nombres = ['Fernando','Gabriel','Diego']
edades = [21,24,23,22]
altura=[120,130,140]
jugadores = zip(nombres,edades, altura)
print(list(jugadores))
```

```python
nombres = ['Fernando','Gabriel','Diego']
edades = [21,24,23,22]
dd= [21,33,36]
jugadores_z = zip(nombres,edades, dd)
# print(list(jugadores_z))
jugadores_uz = zip(*jugadores_z)
print(list(jugadores_uz))
```
