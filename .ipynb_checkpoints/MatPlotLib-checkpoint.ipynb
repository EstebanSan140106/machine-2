### 1.0.1 Elementos de visualización
- Gráfico: Representación dentro de una subfigura o figura. Hay muchos tipos de gráficos: líneas, barras, dispersión, áreas, etc.
- Conceptos clave:
  - Figura: contenedor general.
  - Subfigura (Axes): cada gráfico individual dentro de la figura.

```python
# Generar datos

x_data_1 = np.arange(0,10)
x_data_2 = np.arange(-10, 0)
y_data_1 = np.random.randint(100 , size=(10))
y_data_2 = np.random.randint(1000, size =(10))
```

```python
# Creacion de figura y subplots

fig , axis = plt.subplots(1,2)

axis[0].plot(x_data_1, y_data_1)
axis[1].plot(x_data_2, y_data_2)

plt.show() #se recomiendo uar 'show()' y no 'fig.show()'
```

```python
fig , axes = plt.subplots(1,2)

axes[0].plot(x_data_1 , y_data_1)
axes[1].plot(x_data_2 , y_data_2)

axes[0].set_xlabel("Años")
axes[0].set_ylabel("Dólares")

axes[1].set_ylim([0,1000])
axes[1].set_xlim([-11,1])
axes[1].set_xlabel("Años")
axes[1].set_ylabel("Dólares" , loc='center', labelpad=-210)
plt.show()
```

## 2 Más Métodos
Multples Parámetros

`plt.plot(x, y, style , label, linewidth=1, alpha=1)`
- `style` : Cadena de texto en formato(color, tipo, etc.)
- `label`: Referencia para la leyenda
- `linewidth`: Grosor de linea
- `alpha`: transparencia

```python
y_data = np.random.randint(100, size=(10))
x_data = np.arange(10)

fig, axis = plt.subplots(1,1)
axis.plot(x_data, y_data, 'gh--' , label='dólares' , linewidth=1, alpha=0.6)
axis.legend(loc='best')
plt.show()
```

## 3 Anotaciones

Son formas de añadir nueva información al gráfico <br>
`axes.test(x,y,'texto')` <br>
`axes.annotate(texto, xy,xytext,xycoords, arrowprops)`

```python
y_data = np.random.randint(100,size=(10))
x_data= np.arange(10)

fig, axis = plt.subplots(1,1)

axis.plot(x_data,y_data)

axis.annotate('Low Income' , xy=(x_data[1], y_data[1]), xytext=(5,10), arrowprops = dict(facecolor='red' , shrink=1.0))
plt.show()
```

## 4 Tipos de gráficos

```python
fig, ax = plt.subplots()

personas = ('PhD', 'MSc' , 'BSc', 'N/A')

y_pos = np.arange(len(personas))
porcentaje = [5,45,25,25]

ax.barh(y_pos, porcentaje, align='center' ,
         color='green' , edgecolor='red')
ax.set_yticks(y_pos)
ax.set_yticklabels(personas)

ax.set_xlabel('Porcentaje de Trabajo')
ax.set_title('Nivel de educación en la companía')
plt.show()
```

```python
ls_count = (284487, 244560, 208493, 196764)
cities = ('Madrid', 'Barcelona', 'Sevilla', 'Valencia')
data = np.array(ls_count)/1000
y_pos = np.arange(data.size)
width = 0.35 # the width of the bars
opacity = 0.7
fig, ax = plt.subplots()
rects1 = ax.barh(y_pos, data, width, alpha=opacity)
rects1[3].set_color('#1f77b4')  # azul estándar
rects1[2].set_color('#ff7f0e')  # naranja estándar
rects1[1].set_color('#2ca02c')  # verde estándar
rects1[0].set_color('#d62728')
ax.set_yticks(y_pos)
ax.set_yticklabels(cities)
# add some text for labels, title and axes ticks
ax.set_ylabel('Ciudades')
ax.set_xlabel('Km de carril bici')
ax.legend((rects1[0], rects1[1], rects1[2], rects1[3]), cities)
plt.show()
```

## 6 frame.plot()

`frame.plot(x,y,label=<Series>, ax_style, kind, xticks, yticks, title, subplots)`
- (Exclusivo para Series) label: referencia a leyenda
- x: subfigura en la que dibujar los datos
- style: estilo de la línea
- kind: tipo de gráfico (’bar’, ‘pie’, ‘hist’, ‘area’ ‘line’, ‘barh’, ‘density’, ‘kde’).
- xticks y yticks: valores en los ejes X e Y
- title: string como título
- (Exclusivo para DataFrame) subplots: bool si se desean subfiguras separadas para cada columna.
- (Exclusivo para DataFrame) ‘x’ e ‘y’ se pueden utilizar para seleccionar columnas para cada eje

```python
rand_matrix = np.random.randint(100, size=(10,3))
frame = pd.DataFrame(rand_matrix, columns=list('ABC'))

frame.plot(style=['r-','g-', 'b-'] , xticks=range(-2,12), title='Random Lines', subplots=True)

plt.show()

display(frame)
```

```python
rand_matrix = np.random.randint(100, size=(10,3))
frame = pd.DataFrame(rand_matrix, columns=list('ABC'))
fig, axis = plt.subplots(3,1)
ax = frame.plot(style=['r-','g-','b-'], xticks=range(-2,12), title='Random lines', subplots=True, ax=axis)
ax[0].legend(loc='upper left')
plt.show()
```

```python
iris = sns.load_dataset('iris')
data = []
colors = ["red", "green", "blue"]
groups = ["setosa", "versicolor", "virginica"]
columns = ["petal_length", "petal_width"]
for i in groups:
    data.append(iris[iris["species"] == i][columns].values)

fig = plt.figure()
ax = fig.add_subplot()
for count, d in enumerate(data):
    x, y = d[:,0], d[:,1]
    ax.scatter(x, y, alpha = 0.4, c = colors[count -1], edgecolors = 'none', s= 30, label = groups[count - 1])

ax.set_xlabel(columns[0])
ax.set_ylabel(columns[1])
plt.legend(loc='best')
plt.title('Matplot scatter plot')
plt.show()
```

8.1

```python
import matplotlib.pyplot as plt
import numpy as np

years = np.arange(2000, 2021)

np.random.seed(42)
y0 = np.random.normal(loc=100, scale=15, size=len(years))
y1 = np.random.normal(loc=120, scale=20, size=len(years))
y2 = np.random.normal(loc=90, scale=10, size=len(years))
y3 = np.random.normal(loc=110, scale=18, size=len(years))

plt.figure(figsize=(10, 6))
plt.plot(years, y0, label='Hawaii')
plt.plot(years, y1, label='San Marino')
plt.plot(years, y2, label='Islas Feroe')
plt.plot(years, y3, label='Guayana')

plt.title('Datos aleatorios')
plt.xlabel('Año')
plt.ylabel('GDP')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Datos aleatorios', fontsize=16)

axes[0, 0].plot(years, y0, color='tab:blue')
axes[0, 0].set_title('Hawaii')
axes[0, 0].set_xlabel('Año')
axes[0, 0].set_ylabel('GDP')
axes[0, 0].grid(True)
axes[0, 0].annotate('Pico importante', xy=(2010, y0[10]), xytext=(2005, y0[10] + 20),
                    arrowprops=dict(arrowstyle='->', color='red'), color='red')

axes[0, 1].plot(years, y1, color='tab:orange')
axes[0, 1].set_title('San Marino')
axes[0, 1].set_xlabel('Año')
axes[0, 1].set_ylabel('GDP')
axes[0, 1].grid(True)

axes[1, 0].plot(years, y2, color='tab:green')
axes[1, 0].set_title('Islas Feroe')
axes[1, 0].set_xlabel('Año')
axes[1, 0].set_ylabel('GDP')
axes[1, 0].grid(True)

axes[1, 1].plot(years, y3, color='tab:red')
axes[1, 1].set_title('Guayana')
axes[1, 1].set_xlabel('Año')
axes[1, 1].set_ylabel('GDP')
axes[1, 1].grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
