##  Funciones

### 1.1 ¿ Qué son? 

- Bloques de código reutilizables que reciben entradas (parámetros) y devuelven un resultado.
- Se definen con la palabra clave `def` y se ejecutan al ser llamadas por su nombre.

### 1.2 Ejemplo

`````python
def suma (a,b) :
    suma = a +b
    return suma
print (suma(7,3))
print (suma(-8,7)) 
`````

`````python
c = 30 

print (c*1.793 + 32)

c = 28 
print (c*1.793 + 32)
`````

`````python
def celsius_a_fahrenheit (c) :
    return c * 1.793+32
`````

`````python
grados = 15
print(celsius_a_fahrenheit(grados)) 
`````

### 1.3 Ventajas
- Evitan duplicar código.
- Mejoran la reutilización.

`````python
def clean_strings (strings , ops):
    resultado = []
    for value in strings:
        for function in ops:
            value = function(value)
        resultado.append(value)
    return resultado
    
def concat5(a):
    return a + '_5'

strings = [' valencia ' , '  madrid ', ' barcelona ' ]
print (strings)

operations = [str.strip, str.title , concat5]
print (clean_strings(strings, operations))
`````

### 1.5 Paso de Parámetros

- Tipos simples(inmutables) por valor: int, float, string
    - Elefecto se como si se creara ima copia dentro de la misma funcion, aunque realmente esto no ocurre
    - Los cambios dentro de la función no afectan afura.

`````python
def cambiar_valor(x) :
    x    += 5
    print ('Dentro de la funcion' , x)
    return 
a = 3 
cambiar_valor(a)
print ('Fuera de la funcion' , a)
`````

- Tipos complejos o mutables por referencia : list, set, dictionary
  - Dentro de la funnción se maneja el mismo objeto que se ha pasado desde afuera
  - Los cambios dentro de la función si afecta afuera

`````python
def add_2(x) :
    x.append(4)

lista = [1,5,7,3]

add_2(lista)

print(lista)
`````

`````python
# con tuplas

def add_2 (x): 
    print (id(x))
    return 
    
lista =(0,1)
print (id(lista))
add_2(lista)
print(lista)
`````

- Si quiero modificar un objeto que es de un tipo básico, los cuales se pasan por valor, puedo simplemente devolver el resultado de la función y hacer una asignación

`````python
def mulplicacion_por_2 (a) :
    a *=2
    return a
x = 3
x = mulplicacion_por_2(x)
print(x)
`````

- Si quiero que no se modifique un objeto de un tipo complejo, los cuales se pasan por referencia, puedo pasar una copia a la función

`````python
def add_2(x):
    x.append(2)
    print('Dentro de la función: ' ,x)

lista=[0,1]
add_2(lista.copy())
print ('Fuera de la función: ' , lista)
`````

`````python
# Alternativa para obtener una copia con slicing

lista = [0,1]
add_2(lista[:])
print('Fuera de la función:' , lista)
`````

- Si la lista contiene sublistas anidadas hay que hacer un deep copy

`````python
import copy

lista= [2, 4, 16, 32, [34, 10, [5,5]]]
copia = copy.deepcopy(lista)

copia[0] = 454
copia[4][2][0] = 64

print(lista)
print(copia)

print(f"{id(lista)} - {id(copia)}")
print(f"{id(lista[4])} - {id(copia[4])}")
`````

`````python
def consulta_sql(**kwargs):
    print(kwargs)
    sql = "SELECT * FROM bicis WHERE "
    args = [f"{key}={value}" for key, value in kwargs.items()]
    sql += " AND ".join((args))
    return sql
    
print(consulta_sql(station='alameda', free=10))
`````

### 1.7 Recursividad
- Funciones que se llaman a si mismas
  

`````python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
print(factorial(6))
def fact(n):
    a = 1
    for i in range(1, n+1):
        a *= i
    return a
print(fact(5))
`````

`````python
fact = lambda x: 1 if x == 0 else x * fact(x-1)
fact(7)
`````

## 1.8 Funciones Lambda
- Es posible definir funciones anónimas (sin nombre)
- Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una `def` no puede (por ejemplo dentro de una lista o como parametro de una función)
- Útiles para cuando se necesita pasar una función como argumento a otra
Formato :
` lambda <lista de argumento> : <valor a devolver>`
- Importante: `<valor a devolver> ` no es un conjunto de instrucciones. Simplemente es una expresión return.
- Las funciones definidas con `def`son mas generales:
  - Cualquier cosa que implemente un lambda, se puede implementar como una función convencional con `def`

`````python
# Funcion suma

def suma (x,y,z) :
    return x+y+z
print(suma(3,4,5))

suma2= lambda x,y,z : x+y+z

print(suma2(4,5,1))
`````

`````python
 # La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizar las comparaciones.
# Ejemplo: ordenar strings por número de carácteres.

cities = ['Valencia', 'Tenerife', 'Barcelona', 'Madrid']
cities.sort()
print(cities)

cities.sort(key = lambda x : len(x))
print(cities)

cities.sort(key = len)
print(cities)

def longitud(x):
    return len(x)
    
cities.sort(key = longitud)
print(cities)
`````

`````python
# Mismo ejemplo sin lambda
cities = ['Valencia', 'Tenerife', 'Barcelona', 'Madrid']
def num_caracteres(x):
    return len(x)
    
cities.sort(key = num_caracteres)
print(cities)
`````

- Podemos guardar una referencia para usar repetidamente

`````python
cities = ['Valencia', 'Tenerife', 'Barcelona', 'Madrid']
f = lambda x : len(x)
for s in cities:
    print(f(s))
`````

- Las expresiones lambda pueden formar parte de una lista

`````python
# Lista de lambdas para mostrar las potencias de 2.
pows = [lambda x: x ** 0,
        lambda x: x ** 1,
        lambda x: x ** 2,
        lambda x: x ** 3,
        lambda x: x ** 4]
for f in pows:
    print(f(3))
`````

- Diccionarios se pueden usar como un 'switch' en otros lenguajes
- Ideal para escoger entre varias opciones

`````python
def switch_dict(operator, x, y):
    # crea un diccionario de opciones y selecciona 'operator'
    return {
        'add': lambda: x + y,
        'sub': lambda: x - y,
        'mul': lambda: x * y,
        'div': lambda: x / y,
    }.get(operator, lambda: None)() # retorna None si no encuentra 'operator'
print(switch_dict('add',2,2))
print(switch_dict('div',10,2))
print(switch_dict('mod',17,3))
`````

### 1.9 Scopes 
- La localización de una asignación a una variable en el código determina desde donde puedes acceder a esa variable
- Una variable asignada dentro de una función solo es visible dentro de esa función
- El alcance de la visibilidad de una variable determina su scope
- El termino *scope* hace referencia a un namespace. Por ejemplo, una función establece su propio namespace

`````python
X = 10
def func():
    # X = 20 # Local a la función. Es una variable diferente, no visible fuera de 'func'
    def func_int():
        X = 15
        print(X)
    func_int()
func()
`````

### Reglas LEGB
- Local: variables locales a F.
- Enclosing: variables localizadas en funciones que contienen a (o por encima de) F.
- Global: variables definidas en el módulo (fichero) que no están contenidas en ninguna función.
Este scope no abarca más de un fichero.
- Built-ins: proporcionadas por el lenguaje.

`````python
# Global (X y func)
X = 99
def func(Y): # Local (Y y Z)
    Z = X + Y
    return Z
print(func(3))
`````

`````python
X = 1
def func():
    X = 2 # X es una variable diferente
func()
print(X)
`````

-  La sentencia global nos permite modificar una variable global desde dentro de una función.

`````python
X = 1
def func():
    global X
    X = 2
func()
print(X)
`````

- Solo es necesario usar *global* para referenciar variables

`````python
y, z = 1, 2
def todas_globales():
    global x
    x = y + z # Las tres variables son globales.
todas_globales()
print(x)
print(y)
print(z)
`````

`````python
X = 33
def f1():
    X = 55 # Enclosing scope (para f2)
    def f2():
        print(X)
    f2()
f1()
`````

### Closure
- Las funciones pueden recordar su enclosing scope, independientemente de si estos continuan existiendo o no

`````python
def f1():
    X = 65 # Enclosing scope (para f2)
    def f2():
        print(X)
    return f2 # Devuelve f2, sin invocarla
accion = f1()
accion()
`````

- Sentencia nonlocal se usa para modificar variables que no son locales, pero tampoco globales.

`````python
def f1():
    contador = 5
    def f2():
        nonlocal contador
        contador += 1
        print(contador)
    return f2
    
accion = f1()
accion()
accion()
accion()
`````
